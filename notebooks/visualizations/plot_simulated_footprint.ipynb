```python
from matplotlib import pyplot as plt

from src.common.paths import PlotPaths
from src.inverse_problems.footprints.gaussian_plume_model import GaussianPlumeModel
```

```python
sensor_x, sensor_y= 26, 25

model = GaussianPlumeModel()

num_measurements = 16

footprints = model.get_sensitivities_for_sensor(sensor_x, sensor_y, num_measurements=num_measurements)

last_footprint = footprints[-1]
```

```python
target_dir = PlotPaths.PLOTS / "gaussian_plume_footprints"
target_dir.mkdir(exist_ok=True, parents=True)
```

```python
_, ax = plt.subplots()
im = ax.imshow(last_footprint, cmap="viridis")
plt.colorbar(im, ax=ax, label=r"$\text{ppm} \cdot \text{m}^2 \cdot \text{s} \cdot \mu\text{mol}^{-1}$")

plt.tight_layout()
plt.savefig(target_dir / f"gaussian_plume_footprint.eps", format="eps", bbox_inches="tight")
```
