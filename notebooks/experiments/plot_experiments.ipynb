```python
ZURICH = "zürich"
MUNICH = "munich"
PARIS = "paris"

fine_tuned_data = {
    MUNICH: EvaluationDataReader(ExperimentPath("fine_tuned_munich_no_noise")),
    ZURICH: EvaluationDataReader(ExperimentPath("fine_tuned_zürich_no_noise")),
    PARIS: EvaluationDataReader(ExperimentPath("fine_tuned_paris_no_noise")),
}
```

```python
for city, data in fine_tuned_data.items():
    logger.info(f"Fine-tuning results for {city.capitalize()}")

    for model in ["256", "512", "1024", "2048"]:
        logger.info(f"\tResults for model {model}")
        
        base_relative_error = data.get_mean_relative_error_per_measurements(solver=model)
        fine_tuned_relative_error = data.get_mean_relative_error_per_measurements(solver=f"{model}_{city}")
        logger.info("\t\tRelative Error")
        for m, (b, f) in zip(data.measurements, zip(base_relative_error, fine_tuned_relative_error)):
            logger.info(f"\t\t\t{m}: {f:.2f} ({100 * ((f - b) / b):.2f}%)")

        base_ssim = data.get_mean_ssim_per_measurements(solver=model)
        fine_tuned_ssim = data.get_mean_ssim_per_measurements(solver=f"{model}_{city}")
        logger.info("\t\tSSIM")
        for m, (b, f) in zip(data.measurements, zip(base_ssim, fine_tuned_ssim)):
            logger.info(f"\t\t\t{m}: {f:.2f} ({100 * ((f - b) / b):.2f}%)")
```

# Gaussian Plume SNR Plot

## Munich

```python
experiment_path = ExperimentPath("gaussian_plume_snr")

city = "munich"

least_squares =  polars.read_csv(experiment_path.path / city / "ls.csv")
bp_dwt = polars.read_csv(experiment_path.path / city / "dwt.csv")
bp_dct = polars.read_csv(experiment_path.path / city / "dct.csv")
gen_2048 =  polars.read_csv(experiment_path.path / city / "2048.csv")
gen_2048_munich = polars.read_csv(experiment_path.path / city / "2048_munich.csv")
```

```python
gaussian_plume_snr_plot_path = PlotPaths.PLOTS / "snr_plots"
gaussian_plume_snr_plot_path.mkdir(exist_ok=True, parents=True)
```

```python
_, axs = plt.subplots(1, 1, figsize=(5, 5))

metric = "relative_error"

axs.plot(least_squares["snr_db"], least_squares[metric], label="LS", marker="o")
axs.plot(bp_dwt["snr_db"], bp_dwt[metric], label="SR (DWT)", marker="s")
axs.plot(bp_dct["snr_db"], bp_dct[metric], label="SR (DCT)", marker="^")
axs.plot(gen_2048["snr_db"], gen_2048[metric], label="VAE", marker="d")
axs.plot(gen_2048_munich["snr_db"], gen_2048_munich[metric], label="VAE Fine-tuned", marker="v")

plt.xlabel("SNR [dB]")
plt.ylabel("Relative Error")
plt.legend(loc="best")
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.savefig(gaussian_plume_snr_plot_path / f"{city}_{metric}.eps", format="eps")
```

```python
_, axs = plt.subplots(1, 1, figsize=(5, 5))

metric = "ssim"

axs.plot(least_squares["snr_db"], least_squares[metric], label="LS", marker="o")
axs.plot(bp_dwt["snr_db"], bp_dwt[metric], label="SR (DWT)", marker="s")
axs.plot(bp_dct["snr_db"], bp_dct[metric], label="SR (DCT)", marker="^")
axs.plot(gen_2048["snr_db"], gen_2048[metric], label="VAE", marker="d")
axs.plot(gen_2048_munich["snr_db"], gen_2048_munich[metric], label="VAE Fine-tuned", marker="v")

plt.xlabel("SNR [dB]")
plt.ylabel("SSIM")
plt.legend(loc="best")
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.savefig(gaussian_plume_snr_plot_path / f"{city}_{metric}.eps", format="eps")
```
