```python
from matplotlib import pyplot as plt

from src.common.paths import ExperimentPath
from src.evaluation.evaluation_data_reader import EvaluationDataReader
```

```python
EXPERIMENT_NAME = "latent_dimension"

experiment_path = ExperimentPath(EXPERIMENT_NAME)
```

```python
data_reader = EvaluationDataReader(experiment_path)
```

```python
plt.figure(figsize=(6, 6))

plt.plot(data_reader.measurements, data_reader.get_mean_relative_error_per_measurements(solver="256"), label="Solver 256", marker="o")
plt.plot(data_reader.measurements, data_reader.get_mean_relative_error_per_measurements(solver="512"), label="Solver 512", marker="s")
plt.plot(data_reader.measurements, data_reader.get_mean_relative_error_per_measurements(solver="1024"), label="Solver 1024", marker="^")
plt.plot(data_reader.measurements, data_reader.get_mean_relative_error_per_measurements(solver="2048"), label="Solver 2048", marker="d")

plt.xscale("log")

plt.xlabel("Measurements (log scale)", fontsize=12)
plt.ylabel("Mean Relative Error", fontsize=12)

plt.legend(loc="best")

plt.grid(True, which="both", ls="--", linewidth=0.5)

plt.tight_layout()

plt.savefig(experiment_path.path / "relative_error.png")
```

```python
plt.figure(figsize=(6, 6))

plt.plot(data_reader.measurements, data_reader.get_mean_ssim_per_measurements(solver="256"), label="Solver 256", marker="o")
plt.plot(data_reader.measurements, data_reader.get_mean_ssim_per_measurements(solver="512"), label="Solver 512", marker="s")
plt.plot(data_reader.measurements, data_reader.get_mean_ssim_per_measurements(solver="1024"), label="Solver 1024", marker="^")
plt.plot(data_reader.measurements, data_reader.get_mean_ssim_per_measurements(solver="2048"), label="Solver 2048", marker="d")

plt.xscale("log")

plt.xlabel("Measurements (log scale)", fontsize=12)
plt.ylabel("Mean SSIM", fontsize=12)

plt.legend(loc="best")

plt.grid(True, which="both", ls="--", linewidth=0.5)

plt.tight_layout()

plt.savefig(experiment_path.path / "ssim.png")
```
